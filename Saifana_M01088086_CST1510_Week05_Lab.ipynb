### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
# Opens .txt file in read only mode and stores lines in "lines"
with open("clothing_stock.txt", "r") as f:
    lines = f.readlines()
# Creates empty dictionary, will be used to track number different sizes of items
item_size_count = {}
# Identifying the item and size within aech line, stores the item in dictionary
for i in lines:
    item, size = map(str.strip, i.split(","))
    if item not in item_size_count:
        # Stores every item in dictionary
        item_size_count[item] = {} 

    if size not in item_size_count[item]:
        item_size_count[item][size] = 0  # Stores every size in respective dictionary
    item_size_count[item][size] += 1  # Accounts and adds duplicate for size
# Prints results in a neat, formatted way
for item, sizes in item_size_count.items():
    print(f"{item} sizes:")
    print("*"*20)
    for size, count in sizes.items():
        print(f"[{size}]: {count} available")
    print()
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
# Utilising sets are the most efficient data structure as you can quickly identify common groups with other sets
students = {"John": {"Math", "Physics"}, "Emily": {"Physics", "Biology"}, "Alex": {"Math"}}
# Making functions that join, leave, and display student study groups
def join(x, y):    # User function for specific student to join a certain study group
    if x not in students:    # Adds if student is not already in student set
        students[x] = set()
    students[x].add(y)       # Adds student to group
    print(f"{x} joined {y}.")
# User function for specific student to leave a certain study group
def leave(x, y):
    if x in students and y in students[x]:   # Removes existing members of study group
        students[x].remove(y)
        print(f"{x} left {y}.")
    else:
        print(f"{x} is not in {y}.")
# User function to display study groups of a student
def display(x):
    if x in students:
        groups = students[x]
        print(f"{x} is in: {', '.join(groups)}")
    else:
        print(f"{x} is not enrolled in any groups.")

# User inputs and calling functions
new = input("Enter Student Name: ")
group = input("Enter Study Group to Join: ")
join(new, group)
new2 = input("Enter Student Name: ")
group2 = input("Enter Study Group to Leave: ")
leave(new2, group2)
new3 = input("Enter Student Name: ")
display(new3)
`````

2. Write a program that simulates playlist operations (add, remove, shuffle).
3. Let users shuffle the playlist and display the new order.

`````python
#Using list data structures to easily remove, add, and shuffle songs
import random  
music = ["Song A", "Song B", "Song C", "Song D"]
# User functions to add songs, detects any duplicates
def addmus():
    x = input("Enter Song to Add: \n")
    if x in music:
        print("The song already exists in the playlist.")
    else:
        music.append(x)
        print(f"{x} Added to playlist.")
# User functions to remove songs, detects if song is already in playlist
def remmus():
    x = input("Enter Song to Remove: \n")
    if x in music:
        music.remove(x)
        print(f"{x} Removed from playlist.")
    else:
        print("The song does not exist in the current playlist.")
# User function to shuffle songs and print new order
def shuff():
    random.shuffle(music)
    order = 0
    print("Shuffled Playlist: \n", "*"*20)
    for i in range(len(music)):
        order += 1
        print(f"{order}. {music[i]}")
# User Interface, letting them choose to add, remove, shuffle songs, or exit music player
choice = 0
while choice != 4:
    choice = int(input("Enter corresponding number: \n1. Add Song \n2. Remove Song \n3. Shuffle Playlist \n4. Exit Music Player\n"))
    # Calling functions
    if choice == 1:
        addmus()
    elif choice == 2:
        remmus()
    elif choice == 3:
        shuff()
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
# M01088086 | Saifana Kuradiela Maryam
# Utilising dictionary data structure
expenses = {
    "Food": [25, 40, 15],
    "Entertainment": [50, 20],
    "Travel": [300, 150]
}
# User function to add expenses
def add(x, y):
    if x not in expenses: # Detects whether category exists, if not will add to dictionary 
        expenses[x] = [] 
    expenses[x].append(y)  # Adds expenses amount to category
    print(f"Added {y} AED to '{x}'.") # Prints results for user visibility
# User function to remove an expense
def remove(x, y):
    if x in expenses and y in expenses[x]: # Detects expenses in category to remove
        expenses[x].remove(y)
        print(f"Removed {y} AED from '{x}'.") # Prints results for user visibility
    else:
        print(f"Expense not found in '{x}'.")  # Case where there are no found expenses from user input
# User function for viewing total spending by category
def view_totals():
    print("\nTotal Spending:\n"+"*"*30)
    for category, amounts in expenses.items():  
        total = sum(amounts)
        print(f"(-) {category}: {total} AED")  # Prints total spending for each expense category 
# User function for printing expense summary
def summary():
    print("\nExpense Summary:\n"+"*"*15)
    for category, amounts in expenses.items():
        print(f"\n{category}:")
        for i, amount in enumerate(amounts, 1):
            print(f"(-) {amount} AED")
    view_totals()  # Calling previous function to calculate total

# User UI
UI = 0

while (UI != 4):
    print("Hello user, what would you like to do today?\n", "*"*15)
    UI = int(input("1. Add Expenses \n2. Remove Expenses \n3. View Spending Summary \n4. Exit\n"))
    if (UI == 1):
        category = input("Enter Expense Category: ")
        expense = int(input("Enter Expense: "))
        add(category, expense)
    elif (UI == 2):
        category = input("Enter Expense Category: ")
        expense = int(input("Enter Expense: "))
        remove(category, expense)
    elif (UI == 3):
        summary()
`````
