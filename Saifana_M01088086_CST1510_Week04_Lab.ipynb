1. Ask user for an input of his/her password and store it in a variable called password
Print the password length first then create function validating if the password has more than eight characters. If the password fulfills the above condition and has more then eight character print 'valid password' otherwise print invalid password and ask the user to enter the password again.

````python
# Defining user function "valid" to determine if parameter "x" is more than 8 in str length
def valid(x):
    v = 0
    if x > 8:
        v = 1
    return v     # Returns value v = 0 or 1

v = 0       
while v == 0:    # Sentinel while loop
    # User input for password
    password = input("Enter a valid password: ")
    length = len(password)
    print(f"Password length: {length}")
    v += valid(length)
    if v == 1:          # Conditions for password validity
        print("Valid password.")
    else:
        print("Invalid password. Try again.")
````

4. Create a function (or set of functions ) taking one parameter for a password.

````python
import string      # Importing function from string library

def counter(x):    # Defining user function "counter" to count different chars in a string
    upcount = 0
    lowcount = 0
    digcount = 0
    specount = 0
    for i in x:              # Checks through different chars within the string
        if i.isupper():
            upcount += 1
        if i.islower():
            lowcount += 1
        if i.isdigit():
            digcount += 1
        if i in string.punctuation:
            specount += 1
    # Prints count of different chars
    print(f"Lower case: {lowcount} \nUpper case: {upcount} \nDigits: {digcount} \nSpecial symbols: {specount}")
# Testing calling function with case below:
password = "P@#yn26at^&i5ve"
counter(password)
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
import random
# Defining user function "secret"
def secret(x, y):
    special_chars = "@#$%^&*!"
    spec = random.randint(0,len(special_chars)) # Randomly generates index position from special characters string
    z = x[0:3] + y[len(y)-2:len(y)] + special_chars[spec] # Takes 1st three chars of first parameter, then last two chars of second parameter
    # + Adds randomly generated index position from special char
    return z
# Asks for user input
name = input("Enter name: ")
surname = input("Enter surname: ")
# Printing user results
print(secret(name, surname))
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
# Defining user function
def binaryToDecimal(binaryString):
    # Keeping count of total and power for 2 multiplier
    total = 0
    power = 0
    for i in binaryString[::-1]:     # Converts binary from right side
        total += int(i) * 2**(power)     # Converting Calculations
        power += 1     # Keeping track for 2 multiplier
    return total       # Returns decimal value

convert = input("Enter binary number: ")
print(f"Decimal Value: {binaryToDecimal(convert)}")
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
def binaryToHex(binaryValue):
    # Creating hex string to refer to
    hexavalue = "0123456789ABCDEF"
    hexadecimal = ""
    # Converting binary using previous binary to decimal function
    decimal = binaryToDecimal(binaryValue)
    if decimal == 0:         # Scenario where binary number is 0
        hexadecimal = "0"
        return hexadecimal
    while decimal > 0:    # Calculations to find hexadecimal from decimal number converted from binary number
        remainder = decimal % 16
        hexadecimal = hexavalue[remainder] + hexadecimal     # Format of hexadecimal
        decimal //= 16
    return hexadecimal

binary = input("Enter binary number: ")
# Prints results
print(f"Hexadecimal Number: {binaryToHex(binary)}")
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
def decimalToBinary(value):
    binary = ""
    while value > 0:            # Calculations to find binary number from decimal number
        remainder = value % 2
        binary += str(remainder)
        value //= 2
    return binary[::-1]        # Returns binary number in reverse order

decimal = int(input("Enter decimal value: "))          # Asking for user input
print(f"Binary Number: {decimalToBinary(decimal)}")    # Printing final results
````
