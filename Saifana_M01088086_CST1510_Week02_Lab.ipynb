4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
#Converting standard units to freedom units
k = 0.45359237
m = 0.0254

#User input of personal weight (lbs), and height (inch)
lbs = float(input("Please enter your weight in pounds (lbs): "))
inch = float(input("Please enter your height in inches (in): "))

#Calculates user's BMI and prints the value
BMI = (lbs * k)/(inch * m)**2
print(f'You have a BMI of: {BMI:.2f}')

#Condition for each range and display's user's BMI range
if (BMI < 18.5) :
    print("Your BMI is within the underweight range.")

elif (18.5 <= BMI <= 25.0) :
    print("Your BMI is within the normal range.")

elif (25.0 <= BMI <= 30.0) :
    print("Your BMI is within the overweight range.")

elif (BMI >= 30.0) :
    print("Your BMI is within the obese range.")
````

15. **Find the highest score**

Write a program that prompts the user to enter the number of students and each student’s name and score and display the student name and highest score.


````python
#Asks for user input for how many students in total, n
n = int(input("How many students in total: "))
#Determining lists, for student names and scores
names = []
scores = []

#Defining x and starting from 1
x = 1
#Loop to add names and scores to lists
while (x <= n) :
    stdName = input(f"Student {x}'s name: ")
    names.append(stdName)
    stdScore = int(input(f"Student {x}'s score: "))
    scores.append(stdScore)
    x += 1
#Determining the top score as well as the cell position in the list
ind = scores.index(max(scores))
#Determining the student with the top score and printing the result
for i in names :
    if names.index(i) == ind :
        print(f"Highest Score: \nStudent Name: {i} \nScore: {max(scores)}")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
#Getting the secret number
num = random.randint(0,10)
#Keeping score (tries) as well as a variable (end) to end the while loop
end = 1
tries = 0

#Loop which continuously asks the user for guesses until the condition is met
while end != 0 :
    guess = int(input("Guess the secret number from 0 to 10:"))
    if guess < num :
        print("The number is too small. Try again.")
        tries += 1
    elif guess > num :
        print("The number is too big. Try again.")
        tries += 1
    else :
        tries += 1
        #To signal the end of the loop
        end -= 1
        #Each guess accumulates 1 try and is then printed in the end after the correct guess
        print(f"You took {tries} tries to guess correctly! The secret number was {num}.")
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
#Dictating lists, and a variable (end) to stop the loop
end = 0
numbers = []
even = []

#Setting a condition to stop the loop, and asking the user to keep inputting a positive integer until entering "done"
while end != 1 :
    num = input('Enter a positive number, or type "done" to end the list: ').lower()    #Using .lower() function to ensure no case sensitivity
    if num == "done" :
        end += 1
    else :
        if num.isdigit() == True :                  #Utilising boolean, if input is a number, it will be added to the list
            numbers.append(int(num))
        elif num.isdigit() == False :
            print("Please only enter numbers.")     #If input is anything else but a number, it will be ignored

#Determining the existence of even numbers in the list, numbers
tally = 0
for i in numbers :
    if (i%2 == 0) :
        even.append(i)
        tally += 1
#Sorting the even list of numbers from ascending order, and printing the first even number 
if len(even) > 0 :
    numbers.sort()
    print(numbers)
    even.sort()
    print(f'The first even number of the list is: {even[0]}')
#Scenario where there are no even numbers
if (tally == 0) and (len(numbers) > 0) :
    numbers.sort()
    print(numbers)
    print("No even numbers were found.")
#Scenario where there were no numbers entered
if (len(numbers) == 0) :
    print("No numbers were entered.")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
#Importing functions from random library
import random
#Determining variables with the randomised lottery numbers
lotto_1 = random.randint(0,9)
lotto_2 = random.randint(0,9)
#Adding the numbers to a list
num_lotto = [lotto_1, lotto_2]

#Asking user for their guess in the form of a string
guess = str(input("Enter a two-digit number to guess the winning lotto numbers: "))
#Converting the guess into a string to add to a list, separating the numbers into individual cells
guess_lotto = list(map(int, str(guess)))

#Scenario where user gets position and numbers perfectly
if num_lotto == guess_lotto :
    print(f"The lotto numbers were {lotto_1}{lotto_2}. You have won £10,000!")

#Scenarios that determine 1 win from the number irregardless of position
else:
    win = 0
    for i in guess_lotto :
        if i in num_lotto :
            win += 1
            
#Conditions for amount of wins to determine cash prize or loss
    if win == 2 :
        print(f"The lotto numbers were {lotto_1}{lotto_2}. You have won £3,000!")
    elif win == 1 :
        print(f"The lotto numbers were {lotto_1}{lotto_2}. You have won £1,000!")
    else :
        print(f"The lotto numbers were {lotto_1}{lotto_2}. Try again next time.")
````

#### 20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
#Asks for user input for how many students in total, n
n = int(input("How many students in total: "))
#Determining lists, for student names and scores
names = []
scores = []

#Defining x and starting from 1
x = 1
#Loop to add names and scores to lists
while (x <= n) :
    stdName = input(f"Student {x}'s name: ")
    names.append(stdName)
    stdScore = int(input(f"Student {x}'s score: "))
    scores.append(stdScore)
    x += 1
#Determining the top score as well as the cell position in the list
ind = scores.index(max(scores))
#Determining the student with the top score and printing the result
for i in names :
    if names.index(i) == ind :
        stdHighest = i
        print(f"Highest Score: \nStudent Name: {i} \nScore: {max(scores)}")
#Removing the highest result along with the student's name
names.remove(stdHighest)
scores.remove(max(scores))

#Determining the top score again as well as the cell position in the list
ind2 = scores.index(max(scores))
#Determining the student with the 2nd top score and printing the result
for i in names :
    if names.index(i) == ind2 :
        print(f"Second Highest Score: \nStudent Name: {i} \nScore: {max(scores)}")
````

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters
